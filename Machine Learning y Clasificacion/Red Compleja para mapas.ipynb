```python
# Importar las bibliotecas necesarias
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo
G = nx.Graph()

# Añadir nodos y aristas con sus pesos
edges = [
    (1, 2, {'weight': 0.5, 'distance': 0.3}),
    (2, 3, {'weight': 1.0, 'distance': 0.6}),
    (3, 4, {'weight': 1.5, 'distance': 1.3}),
    (4, 5, {'weight': 1.0, 'distance': 0.5}),
    (4, 7, {'weight': 1.5, 'distance': 0.6}),
    (4, 8, {'weight': 0.5, 'distance': 0.3}),
    (5, 6, {'weight': 1.0, 'distance': 0.7}),
    (6, 7, {'weight': 0.2, 'distance': 0.2}),
    (7, 13, {'weight': 1.5, 'distance': 1.0}),
    (7, 16, {'weight': 1.0, 'distance': 0.5}),
    (8, 10, {'weight': 0.5, 'distance': 0.4}),
    (9, 10, {'weight': 1.0, 'distance': 0.5}),
    (10, 11, {'weight': 1.0, 'distance': 0.5}),
    (10, 12, {'weight': 1.5, 'distance': 1.5}),
    (12, 13, {'weight': 0.5, 'distance': 0.4}),
    (13, 14, {'weight': 1.0, 'distance': 0.5}),
    (14, 15, {'weight': 1.0, 'distance': 0.5}),
    (14, 17, {'weight': 1.5, 'distance': 0.6}),
    (15, 20, {'weight': 0.5, 'distance': 0.4}),
    (16, 17, {'weight': 0.5, 'distance': 0.4}),
    (17, 18, {'weight': 1.0, 'distance': 0.5}),
    (17, 19, {'weight': 0.2, 'distance': 0.3}),
    (19, 20, {'weight': 0.2, 'distance': 0.3})
]

G.add_edges_from(edges)

# Función para visualizar el grafo
def plot_graph(G):
    # Usar un layout con más espacio
    pos = nx.spring_layout(G, k=0.1, iterations=50)

    # Ajustar el tamaño de la figura
    plt.figure(figsize=(10, 7))

    # Dibujar nodos y aristas
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=200, font_size=10, font_weight='bold')

    # Dibujar etiquetas de las aristas con distancias
    edge_labels = nx.get_edge_attributes(G, 'distance')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=7, label_pos=0.5)

    # Configurar el título y eliminar ejes
    plt.title("Grafo de El Florido", fontsize=14)
    plt.axis('off')
    plt.show()


# Función para encontrar la ruta más corta considerando los pesos
def shortest_path(G, start, end):
    path = nx.dijkstra_path(G, start, end, weight='weight')
    length = nx.dijkstra_path_length(G, start, end, weight='weight')
    return path, length

# Visualizar el grafo
plot_graph(G)

# Ejemplo de uso: encontrar la ruta más corta del nodo 1 al 20
start, end = 1, 20
path, length = shortest_path(G, start, end)
print(f"La ruta más corta de {start} a {end} es: {path}")
print(f"El peso total de la ruta es: {length:.2f}")

# Función para encontrar la ruta más corta considerando la distancia
def shortest_path_by_distance(G, start, end):
    path = nx.dijkstra_path(G, start, end, weight='distance')
    length = nx.dijkstra_path_length(G, start, end, weight='distance')
    return path, length

# Ejemplo de uso: encontrar la ruta más corta del nodo 1 al 20 considerando la distancia
start, end = 1, 20
path_distance, length_distance = shortest_path_by_distance(G, start, end)
print(f"\nLa ruta más corta considerando la distancia de {start} a {end} es: {path_distance}")
print(f"La longitud total de la ruta considerando la distancia es: {length_distance:.2f}")
```
