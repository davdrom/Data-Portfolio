```python
import pandas as pd

df = pd.read_csv('diabetes.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar el estilo de los gráficos
sns.set(style="whitegrid")

# Crear histogramas para cada variable
df.hist(figsize=(12, 10), bins=20)
plt.show()
```

```python
# Revisar valores faltantes
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

# Definir las características a escalar
features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Crear un scaler
scaler = StandardScaler()

# Aplicar el escalado a las características
df[features] = scaler.fit_transform(df[features])

# Visualizar el DataFrame escalado
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Crear un normalizador
scaler = MinMaxScaler()

# Aplicar el escalado
df[features] = scaler.fit_transform(df[features])

# Ver el DataFrame normalizado
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Definir las características (X) y la variable objetivo (y)
X = df.drop('Outcome', axis=1)  # Características
y = df['Outcome']               # Variable objetivo

# Dividir el dataset en conjuntos de entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Verificación del tamaño de los conjuntos
print(f"Tamaño del conjunto de entrenamiento: {X_train.shape}")
print(f"Tamaño del conjunto de prueba: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# Crear el modelo Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Entrenar el modelo
rf_model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = rf_model.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo Random Forest: {accuracy:.2f}")

# Reporte de clasificación
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Hacer predicciones en el conjunto de prueba
y_pred = rf_model.predict(X_test)

# 1. Precisión (Accuracy)
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo Random Forest: {accuracy:.2f}")

# 2. Reporte de clasificación (contiene precisión, recall y F1 score)
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred))

# 3. Matriz de Confusión
cm = confusion_matrix(y_test, y_pred)

# Visualizar la matriz de confusión
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Actual')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Obtener la importancia de las características del modelo Random Forest
importances = rf_model.feature_importances_

# Crear un DataFrame para visualizar la importancia
feature_importances = pd.DataFrame({'Variable': X.columns, 'Importancia': importances})

# Ordenar las variables por importancia
feature_importances = feature_importances.sort_values(by='Importancia', ascending=False)

# Mostrar el DataFrame con la importancia de las variables
print(feature_importances)

# Graficar la importancia de las características
plt.figure(figsize=(10, 6))
plt.barh(feature_importances['Variable'], feature_importances['Importancia'], color='skyblue')
plt.title('Importancia de las Características en el Modelo Random Forest')
plt.xlabel('Importancia')
```
