# Técnicas de limpieza de datos:


Manejo de valores faltantes:


Imputación: Reemplazar los valores faltantes con la media, mediana, moda o usando algoritmos de imputación.

Eliminación: Remover filas o columnas con valores faltantes, si no afectan el análisis.

Detección y tratamiento de valores atípicos:

Métodos como el rango intercuartílico (IQR) o z-scores para detectar valores extremos que podrían distorsionar el análisis.

Normalización y estandarización:

Escalar las características para que estén dentro de un rango común, como entre 0 y 1, o hacer que tengan una media de 0 y una desviación estándar de 1.

Eliminación de duplicados:

Identificar y eliminar datos duplicados que puedan distorsionar los resultados del análisis.
Corrección de errores tipográficos:

Revisar datos categóricos para corregir posibles errores de escritura.
# Estadísticas descriptivas:
Medidas de tendencia central:

Media: El promedio de los datos.

Mediana: El valor central en un conjunto de datos ordenados.

Moda: El valor que aparece con mayor frecuencia.

Medidas de dispersión:

Varianza y desviación estándar: Cuánto varían los datos respecto a la media.

Rango: Diferencia entre el valor máximo y el mínimo.

Cuantiles y percentiles:

Dividen los datos en intervalos, siendo el percentil 50 equivalente a la mediana.

Distribución de frecuencias:

Utiliza tablas o gráficos para mostrar cuántas veces ocurre cada valor.

# Visualizaciones de datos:

Gráficos de barras:

Útiles para comparar cantidades entre categorías.

Histogramas:

Muestran la distribución de una variable numérica dividiendo los datos en intervalos.

Diagramas de dispersión (scatter plots):

Usados para explorar relaciones entre dos variables.

Gráficos de caja (boxplots):

Ayudan a visualizar la distribución, la mediana y los valores atípicos de los datos.

Mapas de calor (heatmaps):

Representan datos en una cuadrícula con diferentes colores para mostrar la intensidad de los valores.

Aquí tienes un desglose detallado de las columnas del conjunto de datos junto con sus significados:

* Clasificación: Clasificación del jugador basada en la valoración general (OVR) dentro del grupo FC 25.
* Nombre: El nombre completo del jugador.
*Altura: La altura del jugador.
*Peso: El peso del jugador
*Posición: La posición principal que el jugador juega en el campo
*Posiciones alternativas: Otras posiciones que el jugador es capaz de desempeñar con eficacia.
*Edad: edad del jugador.
*País: País al que representa el jugador en competiciones internacionales.
*Liga: La liga de fútbol en la que juega actualmente el jugador.
*Equipo: El equipo del club al que pertenece el jugador.
*Estilo de juego: Rasgos o tendencias de juego específicos que definen el comportamiento y el conjunto de habilidades del jugador en el campo (por ejemplo, «Paso rápido», «Tiro fino»).
*URL: Enlace al perfil detallado del jugador.

Atributos del jugador
* Aceleración: La capacidad del jugador para alcanzar rápidamente la velocidad máxima.
*Velocidad de sprint: La velocidad máxima que el jugador puede alcanzar al esprintar.
*Posicionamiento: La conciencia y el posicionamiento del jugador en ataque.
*Finalización: capacidad del jugador para convertir las ocasiones de gol en goles.
*Potencia de disparo: fuerza de los disparos a puerta del jugador.
*Disparos lejanos: La precisión y la potencia de los disparos efectuados desde fuera del área penal.
*Voleas: La habilidad del jugador para golpear el balón limpiamente desde el aire.
*Penaltis: La habilidad del jugador para lanzar penaltis.

Pases y visión
* Visión: La habilidad del jugador para realizar pases precisos y ver el desarrollo de las jugadas.
*Cruce: La capacidad de realizar centros precisos desde zonas amplias.
*Precisión en los tiros libres: La precisión del jugador en los tiros libres.
*Pases cortos: La precisión y habilidad en la realización de pases a corta distancia.
*Pase largo: La habilidad para realizar pases precisos de larga distancia.
*Curva: La habilidad del jugador para doblar el balón durante los pases o disparos.

Regate y agilidad
* Regate: Control del balón y habilidad del jugador para maniobrar en espacios reducidos.
*Agilidad: rapidez y fluidez con la que el jugador puede cambiar de dirección.
*Equilibrio: La estabilidad del jugador y su capacidad para mantenerse en pie bajo presión.
*Reacciones: La capacidad de reacción del jugador ante acontecimientos imprevisibles durante el juego.
*Control del balón: cómo controla el jugador el balón en movimiento.
Mentalidad y defensa
*Compostura: Calma del jugador bajo presión.
*Intercepciones: La habilidad del jugador para leer e interceptar pases.
*Precisión de cabeza: La precisión del jugador al intentar cabecear el balón.
Conciencia defensiva (Def Awareness): Posicionamiento y capacidad del jugador para anticipar situaciones defensivas.
*Tackle de pie: La habilidad del jugador para ganar el balón con una entrada de pie.
*Tackle deslizante: La habilidad y precisión de los placajes deslizantes del jugador.

Atributos físicos

* Salto: La habilidad del jugador para saltar alto durante los remates de cabeza o los desafíos.
*Resistencia: resistencia y capacidad del jugador para rendir a un alto nivel durante todo el partido.
*Fuerza: La potencia física del jugador y su capacidad para ganar retos físicos.
*Agresividad: La determinación e intensidad del jugador para ganar retos y duelos.

Habilidades técnicas
* Pie débil: La habilidad del jugador con su pie no dominante (valorado de 1 a 5 estrellas).
*Movimientos de habilidad: La habilidad del jugador para realizar movimientos de regate avanzados (valorados de 1 a 5 estrellas).
*Pie preferido: Indica si el jugador prefiere utilizar el pie izquierdo o el derecho.

Atributos del portero (si procede)

* Salto del portero: La habilidad del portero para lanzarse y realizar paradas.
*GK Manipulación: La habilidad del portero para atrapar o retener el balón.
*Golpeo del portero: La precisión y la potencia de las patadas del guardameta al distribuir el balón.
*Posicionamiento del portero: La habilidad del guardameta para posicionarse eficazmente en situaciones defensivas.
*Reflejos del portero: Rapidez del guardameta para reaccionar a los disparos.

```python
# Extraer solo los centímetros de la columna Height
df['Height_cm'] = df['Height'].str.extract(r'(\d+)cm').astype(float)
```

```python
# Extraer solo los kilogramos de la columna Weight
df['Weight_kg'] = df['Weight'].str.extract(r'(\d+)kg').astype(float)
```

```python
# Eliminar las columnas originales
df.drop(columns=['Height', 'Weight'], inplace=True)
```

```python
# Mostrar las primeras filas para confirmar
df[['Height_cm', 'Weight_kg']].head()
```

Eliminar columnas si más del 50% de sus valores son nulos:

```python
# Eliminar columnas con más del 50% de valores nulos
df = df.dropna(thresh=int(df.shape[0] * 0.5), axis=1)
```

Rellenar con valores por defecto (por ejemplo, 0 para atributos numéricos):



```python
# Rellenar con 0 para columnas numéricas
df.fillna(0, inplace=True)
```

```python
# Verificar si quedan valores nulos
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(df['OVR'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de la Clasificación General (OVR)')
plt.xlabel('Clasificación General (OVR)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df[['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']], palette='Set2')
plt.title('Distribución de las Principales Habilidades de los Jugadores')
plt.ylabel('Valor de la Habilidad')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='OVR', data=df, hue='Age', palette='coolwarm')
plt.title('Relación entre la Edad y la Clasificación General (OVR)')
plt.xlabel('Edad')
plt.ylabel('Clasificación General (OVR)')
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr_matrix = df[['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Matriz de Correlación entre Habilidades de Jugadores')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.jointplot(x='Height_cm', y='Weight_kg', data=df, kind='hex', cmap='Oranges')
plt.title('Relación entre Altura y Peso de los Jugadores')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['League'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Número de Jugadores por Liga')
plt.xlabel('Liga')
plt.ylabel('Cantidad de Jugadores')
plt.show()
```
