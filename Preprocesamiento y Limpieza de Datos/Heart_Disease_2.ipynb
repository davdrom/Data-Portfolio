```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

# Fetch dataset
heart_disease = fetch_ucirepo(id=45)

# Data (as pandas DataFrames)
X = heart_disease.data.features
y = heart_disease.data.targets

# Combine features and targets into a single DataFrame
df = pd.concat([X, y], axis=1)

# Display the first few rows of the DataFrame
```

#**Estandarizacion**

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Selecciona las columnas que deseas estandarizar (todas excepto 'num', que parece ser la columna objetivo)
features = df.drop(columns=['num'])

# Inicializa el escalador
scaler = StandardScaler()

# Aplica la estandarización
scaled_features = scaler.fit_transform(features)

# Crea un nuevo DataFrame con los valores estandarizados
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_scaled['num'] = df['num']

# Muestra los primeros datos estandarizados
print(df_scaled.head())
```

# **Normalizacion**

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Selecciona las columnas que deseas normalizar (todas excepto 'num', que es la columna objetivo)
features = df_scaled.drop(columns=['num'])

# Inicializa el escalador
scaler = MinMaxScaler()

# Aplica la normalización
normalized_features = scaler.fit_transform(features)

# Crea un nuevo DataFrame con los valores normalizados
df_normalized = pd.DataFrame(normalized_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_normalized['num'] = df['num']

# Muestra los primeros datos normalizados
print(df_normalized.head())
```

# **Transformación Logarítmica**

```python
import numpy as np
import pandas as pd

# Selecciona las columnas que deseas transformar (todas excepto 'num')
features = df_normalized.drop(columns=['num'])

# Para evitar problemas con ceros, sumamos una pequeña constante (ej: 1)
features_log = features + 1

# Aplica la transformación logarítmica
log_transformed_features = np.log(features_log)

# Crea un nuevo DataFrame con los valores transformados
df_log_transformed = pd.DataFrame(log_transformed_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_log_transformed['num'] = df['num']

# Muestra los primeros datos transformados
print(df_log_transformed.head())
```

# **Funciones Polinómicas**

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

# Selecciona las columnas que deseas usar para generar las características polinómicas (todas excepto 'num')
features = df_log_transformed.drop(columns=['num'])

# Inicializa el generador de funciones polinómicas (por ejemplo, grado 2)
poly = PolynomialFeatures(degree=2, include_bias=False)

# Aplica la transformación polinómica
poly_features = poly.fit_transform(features)

# Crea un nuevo DataFrame con las nuevas características polinómicas
df_poly = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(features.columns))

# Agrega la columna objetivo de nuevo
df_poly['num'] = df['num']

# Muestra los primeros datos con las características polinómicas
print(df_poly.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Función para graficar las distribuciones de las características
def plot_distributions(data, title):
    plt.figure(figsize=(15, 10))
    for i, column in enumerate(data.columns, 1):
        plt.subplot(4, 4, i)
        sns.histplot(data[column], kde=True)
        plt.title(f'{column}')
    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# Distribución original de las características
plot_distributions(df.drop(columns=['num']), 'Distribución Original de las Características')

# Distribución después de la estandarización
plot_distributions(pd.DataFrame(df_scaled, columns=df.columns[:-1]), 'Distribución después de la Estandarización')

# Distribución después de la normalización
plot_distributions(df_normalized.drop(columns=['num']), 'Distribución después de la Normalización')

# Distribución después de la transformación logarítmica
plot_distributions(df_log_transformed.drop(columns=['num']), 'Distribución después de la Transformación Logarítmica')


```
