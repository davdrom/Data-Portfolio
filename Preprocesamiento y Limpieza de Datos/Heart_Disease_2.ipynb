#**Estandarizacion**

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Selecciona las columnas que deseas estandarizar (todas excepto 'num', que parece ser la columna objetivo)
features = df.drop(columns=['num'])

# Inicializa el escalador
scaler = StandardScaler()

# Aplica la estandarizaciÃ³n
scaled_features = scaler.fit_transform(features)

# Crea un nuevo DataFrame con los valores estandarizados
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_scaled['num'] = df['num']

# Muestra los primeros datos estandarizados
print(df_scaled.head())
```

# **Normalizacion**

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Selecciona las columnas que deseas normalizar (todas excepto 'num', que es la columna objetivo)
features = df_scaled.drop(columns=['num'])

# Inicializa el escalador
scaler = MinMaxScaler()

# Aplica la normalizaciÃ³n
normalized_features = scaler.fit_transform(features)

# Crea un nuevo DataFrame con los valores normalizados
df_normalized = pd.DataFrame(normalized_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_normalized['num'] = df['num']

# Muestra los primeros datos normalizados
print(df_normalized.head())
```

# **TransformaciÃ³n LogarÃ­tmica**

```python
import numpy as np
import pandas as pd

# Selecciona las columnas que deseas transformar (todas excepto 'num')
features = df_normalized.drop(columns=['num'])

# Para evitar problemas con ceros, sumamos una pequeÃ±a constante (ej: 1)
features_log = features + 1

# Aplica la transformaciÃ³n logarÃ­tmica
log_transformed_features = np.log(features_log)

# Crea un nuevo DataFrame con los valores transformados
df_log_transformed = pd.DataFrame(log_transformed_features, columns=features.columns)

# Agrega la columna objetivo de nuevo
df_log_transformed['num'] = df['num']

# Muestra los primeros datos transformados
print(df_log_transformed.head())
```

# **Funciones PolinÃ³micas**

```python
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd

# Selecciona las columnas que deseas usar para generar las caracterÃ­sticas polinÃ³micas (todas excepto 'num')
features = df_log_transformed.drop(columns=['num'])

# Inicializa el generador de funciones polinÃ³micas (por ejemplo, grado 2)
poly = PolynomialFeatures(degree=2, include_bias=False)

# Aplica la transformaciÃ³n polinÃ³mica
poly_features = poly.fit_transform(features)

# Crea un nuevo DataFrame con las nuevas caracterÃ­sticas polinÃ³micas
df_poly = pd.DataFrame(poly_features, columns=poly.get_feature_names_out(features.columns))

# Agrega la columna objetivo de nuevo
df_poly['num'] = df['num']

# Muestra los primeros datos con las caracterÃ­sticas polinÃ³micas
print(df_poly.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# FunciÃ³n para graficar las distribuciones de las caracterÃ­sticas
def plot_distributions(data, title):
    plt.figure(figsize=(15, 10))
    for i, column in enumerate(data.columns, 1):
        plt.subplot(4, 4, i)
        sns.histplot(data[column], kde=True)
        plt.title(f'{column}')
    plt.suptitle(title, fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# DistribuciÃ³n original de las caracterÃ­sticas
plot_distributions(df.drop(columns=['num']), 'DistribuciÃ³n Original de las CaracterÃ­sticas')

# DistribuciÃ³n despuÃ©s de la estandarizaciÃ³n
plot_distributions(pd.DataFrame(df_scaled, columns=df.columns[:-1]), 'DistribuciÃ³n despuÃ©s de la EstandarizaciÃ³n')

# DistribuciÃ³n despuÃ©s de la normalizaciÃ³n
plot_distributions(df_normalized.drop(columns=['num']), 'DistribuciÃ³n despuÃ©s de la NormalizaciÃ³n')

# DistribuciÃ³n despuÃ©s de la transformaciÃ³n logarÃ­tmica
plot_distributions(df_log_transformed.drop(columns=['num']), 'DistribuciÃ³n despuÃ©s de la TransformaciÃ³n LogarÃ­tmica')


```
