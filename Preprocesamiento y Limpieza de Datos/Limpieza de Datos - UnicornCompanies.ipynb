```python
# Asegurarse de que los valores sean cadenas y manejar nulos
df['Valuation ($B)'] = df['Valuation ($B)'].fillna('').astype(str).str.replace('[$B]', '', regex=True)

# Convertir a float
df['Valuation ($B)'] = pd.to_numeric(df['Valuation ($B)'], errors='coerce')

# Hacer lo mismo para 'Total Raised'
df['Total Raised'] = df['Total Raised'].fillna('').astype(str).str.replace('[$B]', '', regex=True)
df['Total Raised'] = pd.to_numeric(df['Total Raised'], errors='coerce')
df.head()
```
