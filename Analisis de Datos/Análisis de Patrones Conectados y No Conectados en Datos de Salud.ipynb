```python
import pandas as pd

# Reemplaza 'nombre_del_archivo.csv' con el nombre de tu archivo
df = pd.read_csv('diabetic_data.csv')
df.head()
```

```python
df.info()
```

## **Parte 1: Patrones Conectados**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que ya cargaste el DataFrame como diabeticdata
# Selección de las variables
variables_seleccionadas = df[['time_in_hospital', 'num_medications']]

# Cálculo de la matriz de correlación
correlation_matrix = variables_seleccionadas.corr()

# Visualización de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()

# Gráfica de la relación entre las dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='time_in_hospital', y='num_medications', data=df)
plt.title('Relación entre tiempo en el hospital y número de medicamentos')
plt.xlabel('Tiempo en el hospital (días)')
plt.ylabel('Número de medicamentos')
plt.show()
```

# **Parte 2: Patrones No Conectados**

```python
# Selección de las variables que probablemente no estén conectadas
variables_no_relacionadas = df[['admission_type_id', 'number_outpatient']]

# Cálculo de la matriz de correlación
correlation_matrix_no_relacionadas = variables_no_relacionadas.corr()

# Visualización de la matriz de correlación
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_no_relacionadas, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación de Variables No Relacionadas')
plt.show()

# Gráfica de la relación entre las dos variables con un diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x='admission_type_id', y='number_outpatient', data=df)
plt.title('Relación entre tipo de admisión y número de visitas ambulatorias')
plt.xlabel('Tipo de admisión')
plt.ylabel('Número de visitas ambulatorias')
plt.show()
```

# **Parte 3: Análisis y Reflexión**
## **Preguntas de Análisis:**
**¿Qué observas en la matriz de correlación de los patrones conectados?**
Se observa que existe una correlación positiva moderada de 0.47. Esto sugiere que, a medida que aumenta el tiempo en el hospital, también tiende a aumentar el número de medicamentos administrados.

**¿Cómo se visualiza la relación entre las variables conectadas en el gráfico de dispersión?**
Se observa que, en general, a medida que aumenta el tiempo en el hospital, el número de medicamentos tiende a incrementarse. Aunque hay cierta dispersión en los datos, la tendencia general es que estas dos variables están positivamente relacionadas.

**¿Qué observas en la matriz de correlación de los patrones no conectados?**
Se observa una correlación muy baja de 0.03, lo que indica que estas variables prácticamente no están relacionadas. Esto respalda la hipótesis de que no existe una relación significativa entre el tipo de admisión y el número de visitas ambulatorias.

**¿Cómo se visualiza la relación entre las variables no conectadas en el gráfico de dispersión?**
Se observa que los puntos en el gráfico de dispersión estan distribuidos de manera aleatoria sin una tendencia discernible.

En el contexto de salud, como en el diagnóstico y tratamiento de la enfermedades, identificar patrones conectados y no conectados entre variables es crucial para tomar decisiones informadas. Los patrones conectados, como la correlación entre el tiempo de hospitalización y el número de medicamentos administrados, pueden indicar relaciones causales o tendencias que ayudan a personalizar y optimizar los tratamientos para mejorar los resultados del paciente. Por otro lado, reconocer patrones no conectados, como la falta de relación entre el tipo de admisión y el número de visitas ambulatorias, evita malinterpretaciones que podrían llevar a tratamientos innecesarios o mal dirigidos. En conjunto, entender estas conexiones o la falta de ellas permite a los profesionales de la salud basar sus decisiones en evidencia sólida, lo que resulta en una atención más efectiva y eficiente.
