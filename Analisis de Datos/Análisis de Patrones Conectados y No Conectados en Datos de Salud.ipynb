```python
import pandas as pd

# Reemplaza 'nombre_del_archivo.csv' con el nombre de tu archivo
df = pd.read_csv('diabetic_data.csv')
df.head()
```

```python
df.info()
```

## **Parte 1: Patrones Conectados**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que ya cargaste el DataFrame como diabeticdata
# Seleccin de las variables
variables_seleccionadas = df[['time_in_hospital', 'num_medications']]

# Clculo de la matriz de correlacin
correlation_matrix = variables_seleccionadas.corr()

# Visualizacin de la matriz de correlacin
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlacin')
plt.show()

# Grfica de la relacin entre las dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='time_in_hospital', y='num_medications', data=df)
plt.title('Relacin entre tiempo en el hospital y nmero de medicamentos')
plt.xlabel('Tiempo en el hospital (das)')
plt.ylabel('Nmero de medicamentos')
plt.show()
```

# **Parte 2: Patrones No Conectados**

```python
# Seleccin de las variables que probablemente no estn conectadas
variables_no_relacionadas = df[['admission_type_id', 'number_outpatient']]

# Clculo de la matriz de correlacin
correlation_matrix_no_relacionadas = variables_no_relacionadas.corr()

# Visualizacin de la matriz de correlacin
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_no_relacionadas, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlacin de Variables No Relacionadas')
plt.show()

# Grfica de la relacin entre las dos variables con un diagrama de dispersin
plt.figure(figsize=(10, 6))
sns.scatterplot(x='admission_type_id', y='number_outpatient', data=df)
plt.title('Relacin entre tipo de admisin y nmero de visitas ambulatorias')
plt.xlabel('Tipo de admisin')
plt.ylabel('Nmero de visitas ambulatorias')
plt.show()
```
